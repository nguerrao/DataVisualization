## Part 1: Global emissions

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
global_emissions = pd.read_csv("./datasets/historical_global_emissions.csv")
```

**1. Displaying raw data**

```python
global_emissions
```

**2. Removing unecessary columns**

Some columns won't be useful for our analysis. Indeed, we are interested in the most recent emissions statistics.


```python
print(global_emissions['Data source'].unique())
print(global_emissions['Gas'].unique())
print(global_emissions['Unit'].unique())
```

```python
global_emissions_filtered = global_emissions[['Country', 'Sector', '2017']]
```

```python
global_emissions_filtered.head(14)
```

```python
# Let's see if we have all the sectors displayed.
print(global_emissions_filtered['Sector'].unique())
```

In order to reduce the complexity of the dataset, we are going to use only 5 sectors:

Here are the sectors provided in this dataset and how they are computed:
"Main IPCC sectors, including energy sub-sectors. Includes:
- agriculture
- bunker fuels
- energy (sums of)
    - electricity/heat
    - fugitive emissions
    - manufacturing/construction
    - building
    - other fuel combustion
    - transportation
- industrial processes
- land-use change and forestry
- waste


We will make 5 sectors from the above one to reduce the complexity:
- electricity/energy (electricity/heat + building)
- industry (industrial processes + manufacturing/construction + fugitive emissions)
- transport (transportation + bunker fuels - used in boats -)
- agriculture and food (agriculture + land-use change and forestry + waste)
- other (other fuel combustion)

We will also remove the sectors that are sums of others (totals and energy)

```python
data = {}
from collections import defaultdict

for index, row in global_emissions_filtered.iterrows():
    
    if row['Country'] not in data:
        data[row['Country']] = defaultdict(lambda:0)
    
    country_data = data[row['Country']]
    
    country_data[row['Sector']] = row['2017']
    
    
rows = []

for key, val in data.items():
    
    country_data = []
    country_data.append(key)
    
    energy = val['Electricity/Heat'] + val['Building']
    industry = val['Industrial Processes'] + val['Manufacturing/Construction'] + val['Fugitive Emissions']
    transport = val['Transportation'] + val['Bunker Fuels']
    agriculture = val['Agriculture'] + val['Land-Use Change and Forestry'] + val['Waste']
    other =  val['Other Fuel Combustion']
    
    country_data.append(energy)
    country_data.append(industry)
    country_data.append(transport)
    country_data.append(agriculture)
    country_data.append(other)
    
    rows.append(country_data)
```

```python
df = pd.DataFrame(rows, columns=['Country', 'Energy', 'Industry', 'Transport', 'Agriculture', 'Other'])
df
```

**3. Analysis**

We can now analyse this dataset.
For each sector we have the MtCo2e (Million tons of carbon dioxide equivalent) for all GHB (green house gas) gases.

We will first show a ranking of country and then find which sector contributes the most to emissions.

```python
df['Total'] = df['Energy'] + df['Industry'] + df['Transport'] + df['Agriculture'] + df['Other']
```

```python
df
```

```python
result = df.sort_values(['Total'], ascending = False)[['Country', 'Total']].head(50)
```

```python
result.head(20)
```

```python
fig = plt.figure(dpi=220)
plt.title("Emissions in MtCo2e per countries in 2017")
plt.rcParams.update({'font.size': 12})
plt.ylabel("Emissions in MtCo2e")
ax = result.plot.bar(x = 'Country', y ='Total', rot = 90, figsize=(30,15), ax=plt.gca(), color = "lightblue")
```

We can see that the world emitted ***51 billion*** tons of Co2 equivalent gases in 2017.
Moreover, it is highly disproportionate. Indeed, China and the USA seem to pollute the most by far. What is interesting to note is that the whole European Union pollute as much as India. This means that with 450 million inhabitants, the EU has the same contribution as a country with 1.2 billion people. 

Let's now look at sectors.

```python
energy = df['Energy'].sum()
industry = df['Industry'].sum()
transport = df['Transport'].sum()
agriculture = df['Agriculture'].sum() 
other = df['Other'].sum()

total = energy + industry + transport + agriculture
energy_p = energy / total
industry_p = industry / total
transport_p = transport / total
agriculture_p = agriculture / total
other_p = other / total
```

```python
fig = plt.figure(figsize=(15, 8), dpi=150)
plt.title("% of emissions by sectors")
plt.xlabel("Sector")
plt.ylabel("% of total emissions")
plt.bar(x = ['Energy', 'Industry', 'Agriculture', 'Transport', 'Other'], 
   height = [energy_p, industry_p, agriculture_p, transport_p, other_p], color = "yellowgreen")
plt.show()
```

The Energy sector represents more than 35% of total emissions. However, it might be understandable as electricity is then used by all the sectors.

Surprisingly, the transportation sector emits less than both the agriculture and industrial ones.

## Part 2: Energy sector

We will now look more precisely at the energy sector. What sources of energy are we using?

```python
global_electricity_sources = pd.read_csv("./datasets/electricity_emissions.csv")
```

```python
# Just take the 2019 Values, not all countries have the 2020 values
# 71 countries have 2020 values, 235 (all countries) have 2019 values
global_electricity_sources_filtered = global_electricity_sources[global_electricity_sources['Year'] == 2019]
```

```python
global_electricity_sources_filtered.head()
```

```python
total_coal = global_electricity_sources_filtered['Electricity from coal (TWh)'].sum()
total_gas = global_electricity_sources_filtered['Electricity from gas (TWh)'].sum()
total_renewables = global_electricity_sources_filtered['Electricity from other renewables (TWh)'].sum()
total_solar = global_electricity_sources_filtered['Electricity from solar (TWh)'].sum()
total_oil = global_electricity_sources_filtered['Electricity from oil (TWh)'].sum()
total_wind = global_electricity_sources_filtered['Electricity from wind (TWh)'].sum()
total_nuclear = global_electricity_sources_filtered['Electricity from nuclear (TWh)'].sum()
```

```python
energy_sector = pd.DataFrame([[total_coal, total_gas, total_renewables, total_solar, total_oil, total_wind, total_nuclear]],columns=['Coal', 'Gas', 'Other renewables', 'Solar', 'Oil', 'Wind', 'Nuclear'])
energy_sector
```

```python
fig = plt.figure(figsize=(7, 7), dpi = 150)

plt.pie(energy_sector.values.squeeze(), labels = energy_sector.columns, autopct='%1.1f%%')
plt.rcParams.update({'font.size': 8})
plt.title("% of total energy for each source")
plt.show()
```

Energies that rely on burning fossil fuel count for more than 70% of the total energy produced.

## Part 3: Industrial sector

For this part, we looked at the emissions of Co2 when manufacturing various products.

```python
industrial_emissions = pd.read_csv("./datasets/industry_emissions.csv", delimiter = ";")
industrial_emissions.head(5)
```

We only keep the most recent data (2016) and rename it.

```python
import warnings
warnings.filterwarnings('ignore')

industrial_emissions_filtered = industrial_emissions[['Materials', '2016']]
# convert value to float, then changing from giga to million of tons of Co2 equivale 
industrial_emissions_filtered['2016'] = industrial_emissions_filtered['2016'].apply(lambda x: float(x.replace(",", ".")) * 1000)
industrial_emissions_filtered.columns = ["Materials", "Total"]
```

```python
sorted_industrial_emissions = industrial_emissions_filtered.sort_values(['Total'], ascending = False)
sorted_industrial_emissions
```

```python
fig = plt.figure(figsize=(15, 8), dpi=150)
plt.title("Emissions due to materials fabrication in 2016")
plt.xlabel("Material")
plt.ylabel("Total emissions (in Million tons of Co2 equivalent)")
plt.bar(x = sorted_industrial_emissions.Materials, height = sorted_industrial_emissions.Total, color = "lightsteelblue")
plt.xticks(rotation=45)
plt.show()
```

Iron & Steel and Cement are materials that pollute a lot. The issue, is that we don't have a green alternative yet. The process itself emits Co2.
